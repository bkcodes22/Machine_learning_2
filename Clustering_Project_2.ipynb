```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Load the data from the CSV file
data = pd.read_excel('C:/Users/manju/Downloads/creditcard1.xlsx')

# Select two columns for clustering
selected_columns = ['CREDIT_LIMIT', 'PAYMENTS']

# Extract the selected columns from the dataframe
X = data[selected_columns]

# Plot the original data
plt.scatter(X[selected_columns[0]], X[selected_columns[1]])
plt.xlabel(selected_columns[0])
plt.ylabel(selected_columns[1])
plt.title('Original Data')

# Perform K-Means clustering with 2 clusters
kmeans = KMeans(n_clusters=2, max_iter=300, tol=1e-5)
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the clustered data with different colors for each cluster and different colors for centroids
plt.scatter(X[selected_columns[0]], X[selected_columns[1]], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200)  # Use red color for centroids

# Generate 10 random data points within the range of the clustered data
min_credit_limit, max_credit_limit = X[selected_columns[0]].min(), X[selected_columns[0]].max()
min_payments, max_payments = X[selected_columns[1]].min(), X[selected_columns[1]].max()

# Generate 10 random data points
random_credit_limits = np.random.uniform(min_credit_limit, max_credit_limit, 10)
random_payments = np.random.uniform(min_payments, max_payments, 10)

# Combine the generated data points into a DataFrame
test_data = pd.DataFrame({
    'CREDIT_LIMIT': random_credit_limits,
    'PAYMENTS': random_payments
})

# Plot the test data points
plt.scatter(test_data['CREDIT_LIMIT'], test_data['PAYMENTS'], label='Test Data', marker='x', c='blue', s=100)

plt.xlabel(selected_columns[0])
plt.ylabel(selected_columns[1])
plt.title('Clustered Data with Test Data')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Elbow Method
inertia = []
for i in range(1, 11):  # You can adjust the range as needed
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data_new)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method graph
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Load the data from the Excel file
data = pd.read_excel('C:/Users/manju/Downloads/creditcard1.xlsx')
```

```python
# Select two columns for clustering
selected_columns = ['CREDIT_LIMIT', 'PAYMENTS']
```

```python
# Extract the selected columns from the dataframe
X = data[selected_columns]
```

```python
# Initialize parameters for analysis
num_clusters_list = [2, 3, 4]  # Try different numbers of clusters
init_methods = ['k-means++', 'random']  # Try different initialization methods
max_iter_list = [100, 300, 500]  # Try different maximum number of iterations
```

```python
# Perform analysis by varying the number of clusters, initialization, and max iterations
for num_clusters in num_clusters_list:
    for init_method in init_methods:
        for max_iter in max_iter_list:
            # Perform K-Means clustering
            kmeans = KMeans(n_clusters=num_clusters, init=init_method, max_iter=max_iter, tol=1e-5, random_state=42)
            kmeans.fit(X)

            # Get cluster labels and centroids
            labels = kmeans.labels_
            centroids = kmeans.cluster_centers_

            # Plot the clustered data with different colors for each cluster and different colors for centroids
            plt.scatter(X[selected_columns[0]], X[selected_columns[1]], c=labels, cmap='viridis')
            plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200)  # Use red color for centroids
            plt.xlabel(selected_columns[0])
            plt.ylabel(selected_columns[1])
            plt.title(f'K-Means Clustering - {num_clusters} clusters, {init_method} initialization, {max_iter} iterations')
            plt.show()

            # Print relevant information
            print(f"Number of clusters: {num_clusters}")
            print(f"Initialization method: {init_method}")
            print(f"Maximum number of iterations: {max_iter}")
            print(f"Final cluster centers: {centroids}")
            print("\n")
```
